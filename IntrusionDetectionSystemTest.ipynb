```python
import pandas as pd
df = pd.read_csv('/drive/My Drive/CIP/TestDataset.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0'],axis = 1, inplace=True)
df.drop(['outcome'],axis = 1, inplace=True)
```

```python
df
```

```python
def label_class(x):
  if x == 'normal.':
    return 0
  elif x == 'DoS':
    return 1
  elif x == 'Probe':
    return 2
  elif x == 'R2L':
    return 3
  elif x == 'U2R':
    return 4
  
# df['label'] = df['outcome']
df['label'] = df['label'].apply(label_class)
```

```python
df['label'].value_counts()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
from sklearn.preprocessing import StandardScaler
def feature_scaling(X):
  scaler = StandardScaler()
  scaler1 = scaler.fit_transform(X)
 
  return scaler1

feature_scaling(X)
```

```python
import numpy as np
import keras
import os 
import glob
from keras.models import load_model
from keras import backend as K
from sklearn import metrics as m
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn import metrics as m
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

sns.set_style('darkgrid')
```

```python
model= load_model('/drive/MyDrive/CIP/multiclassmodel.h5')
df['label'].value_counts()
y = df.iloc[:,-1]
```

```python
y
```

```python
y_pred = np.argmax(model.predict(X), axis=-1)

print("\n\nPERFORMANCE OF MODEL ON TESTING DATA")
print("Accuracy: ",accuracy_score(y,y_pred))
confusionmatrix = m.confusion_matrix(y, y_pred)
print("\nConfusion Matrix:\n", confusionmatrix)
print("\n\n")


figure = plt.figure(figsize=(6, 5))
sns.heatmap(confusionmatrix, annot=True, fmt="d")
plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.show()

from sklearn.metrics import classification_report
print('\n\nClassification Report\n',classification_report(y, y_pred,zero_division = 1 ))
```

```python
print("Average f1 score: ",sklearn.metrics.f1_score(y,y_pred,average='weighted'))
print("Average Accuracy score: ",sklearn.metrics.accuracy_score(y,y_pred))
print("Average precision score: ",sklearn.metrics.precision_score(y,y_pred,average='weighted',zero_division = 1))
print("Average precision score: ",sklearn.metrics.recall_score(y,y_pred,zero_division = 1,average='weighted'))
```
